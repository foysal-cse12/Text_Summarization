```python
import json
import gzip
import os
import sys
import argparse
import pandas as pd
```

```python
publication_number = []
description = []
summary = []
    
def readData(input_path,split_type,cpc_code):
    
    
    file_names = os.listdir(os.path.join(input_path,split_type,cpc_code))
    #print('file_names: ',file_names)
    print('total number of file: ',len(file_names))
    # reading one of the gz files.
    for i in range(len(file_names)):
        
        file_name = file_names[i]
        print("Reading file "+ file_name + " from "+ split_type+" split for cpc code " + cpc_code)
        
        with gzip.open(os.path.join(input_path,split_type,cpc_code,file_name),'r') as fin:
            for row in fin:
                json_obj = json.loads(row)
                #print("Fields present in each of the json object:")
                #print(json_obj.keys())
                # uncomment to print the publication number, abstract and description
                print(json_obj["publication_number"])
                # print(json_obj["abstract"])
                # print(json_obj["description"])
               # break
                publication_number.append(json_obj["publication_number"])
                description.append(json_obj["description"])
                summary.append(json_obj["abstract"])
              
    
#readData('G:\\master thesis\\big patent dtaset\\bigPatentData','test','a')     
readData('G:\\master thesis\\big patent dtaset\\bigPatentData','val','e')  
    
```

```python
print(len(publication_number)) 
print(len(description)) 
print(len(summary)) 
```

```python
# https://cmdlinetips.com/2018/01/how-to-create-pandas-dataframe-from-multiple-lists/
data = {'Publication_number':publication_number,'Description':description,'Summary':summary}
df = pd.DataFrame(data)
df.head(5)
```

## Save as csv file 

```python
df.to_csv('G:\\master thesis\\bigPatent_dataset_csv\\val\\val_E.csv', encoding='utf-8', index=0)
print('csv file generated')
```
